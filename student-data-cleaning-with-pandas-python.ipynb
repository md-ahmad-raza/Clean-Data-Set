```python
import pandas as pd
```

```python
# read data set
df = pd.read_csv("Unclean data-set/students_data.csv")
```

```python
# show data set
print(df)
```

```python
# show information and summary about data set
df.info()
```

```python
# check duplicate values
#if return false => duplicate value is not there
#if return true => duplicate value is there
df.duplicated()
```

```python
# remove duplicate value
df.drop_duplicates()
```

```python
# save  data set after remove
df = df.drop_duplicates()
print(df)
```

```python
# check null values
# if return false => there is no null value
# if return true => there is null value
df.isnull()
```

```python
# count total null values
df.isnull().sum()
```

Now here clean data column by column

```python
#for name column 
# this is un structure name column
df["name"]
```

```python
# do capital letter of whole name
df["name"].str.lower().str.title()
```

```python
# update this modified column in original data set
df.loc[:, "name"] = df["name"].str.lower().str.title()
print(df)
```

```python
# for age column
df["age"]
```

```python
# check total null values
df["age"].isnull().sum()
```

```python
# show unique value
df["age"].unique()
```

```python
# find out mean of age
age_mean = df["age"].mean().round() # round() converted decimal value into zero
print(age_mean)
```

```python
# fill mean value instead of NaN value
df["age"].fillna(age_mean)
```

```python
# update age column in original data set
df.loc[:, "age"] = df["age"].fillna(age_mean)
print(df)
```

```python
# for gender column
df["gender"]
```

```python
#check total null value
df["gender"].isnull().sum()
```

```python
#check unique value for understand
df["gender"].unique()
```

```python
# do first letter capital of gender column
df["gender"].str.lower().str.title()
```

```python
#update this column in original data set
df.loc[:, "gender"] = df["gender"].str.lower().str.title()
print(df)
```

```python
# for grade column
df["grade"]
```

```python
#check unique value
df["grade"].unique()
```

```python
# replace 10th to 10
df["grade"].replace({'10th': '10', '11th': '11', '12th': '12'})
```

```python
#update this column in original data set
df.loc[:, "grade"] = df["grade"].replace({'10th': '10', '11th': '11', '12th': '12'})
print(df)
```

```python
# for math_score clumn
df["math_score"]
```

```python
# check total null values
df["math_score"].isnull().sum()
```

```python
# show unique value
df["math_score"].unique()
```

```python
#find out mean
math_mean = df["math_score"].mean()
print(math_mean)
```

```python
# fill math_mean value instead of NaN value
df["math_score"].fillna(math_mean)
```

```python
#update this column into original data set
df.loc[:, "math_score"] = df["math_score"].fillna(math_mean)
print(df)
```

```python
# for english_score column
df["english_score"]
```

```python
#check total null value
df["english_score"].isnull().sum()
```

```python
# show unique value
df["english_score"].unique()
```

```python
# fill eng_mean value instead of NaN
df["english_score"].fillna(eng_mean)
```

```python
# update this column into original data set
df.loc[:, "english_score"] = df["english_score"].fillna(eng_mean)
print(df)
```
